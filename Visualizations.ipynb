```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.size'] = 20

labels = ['LAGB < 15°', 'HAGB > 15°']
shape_ab=[16.87,83.13]
shape_tb=[18.26,81.74]


x = np.arange(len(labels))  # the label locations

width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, shape_ab, width, label='Inc625',color='gray',edgecolor='k')
rects2 = ax.bar(x + width/2, shape_tb, width, label='Inc625 + 0.56 wt.% TiB2',color='steelblue',edgecolor='k')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
# ax.set_xticks(x, labels)

ax.legend(loc='upper left',
          ncol=1, fancybox=True, shadow=True, fontsize=18)

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., h+0.001, "%.2f" %h,
                ha='center', va='bottom',fontsize=18)

autolabel(rects1)
autolabel(rects2)
plt.ylim(0,100)
plt.xticks(x,labels)
plt.ylabel('Relative frequency')
fig.tight_layout()

plt.show()

```
